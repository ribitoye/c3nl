## Types of data

Data in MATLAB can be declared as variables in the form of numbers or strings and these can be arranged into arrays or matrices. Variables are fundamental to all programming; therefore, we will now look at some of the most common variable types.

### Number variables
You can declare a number variable as follows.

```matlab
vara = 1
```

It is possible to define multiple variables at the same time.

```matlab
vara = 7; varb=3;
```

Using ';' suppresses MATLAB from outputting the variables to the command line. The code is still executed though and the number is added to the workspace. This is useful when running more complex code, othewise the output can be overly verbose. You can see the variables that are in the workspace via the Workspace panel. Additionally, you can output their names to the command line panel using the 'who' command.

```matlab
who
```

Typing the variable name will output its value to the comamnd line. Try the following commands.

```matlab
vara
varb
vara + varb
```

By default, MATLAB truncates numbers to four decimal places. You can change the number of decimal places using the 'format' command. Try typing in the following commands, which divide vara by varb.

```matlab
vara / varb

format long;

vara / varb

format short e;

vara / varb

format bank;

vara / varb

format short;

vara / varb
```

Numbers may also be rounded to integers using the int32 command.

```matlab
int32(vara / varb)
```

### Creating vectors
In MATLAB numbers are often stored as vectors or matrices. Vectors can be in vertical and horizontal orientations. You declare vectors using square brakets. Try the following

```matlab
varc = [2 4 6]
```

```matlab
varc = [2; 4; 6]
```

This produces vectors in horizontal and vertical orientations. You can transpose between these orientations.



```matlab
varc = [2 4 6]
varc = varc'
```

Vectors can be added to each other or to individual numbers.



```matlab
varc + vara
vard = [1; 2; 3]
varc + vard
```

However, in order to add vectors they must have the same length.

```matlab
vard = [1; 2; 3; 4]
varc + vard
```

### Creating Strings
Text is stored and manipulated in string variables. When strings are defined they must be enclosed in apostrophe marks, i.e., 'STRING'. Try the following statement.



```matlab
stringa = 'My String'
```

You can use the 'whos' command to examine the properties of this new string variable.




```matlab
whos
```

Here you can see that the string has 9 characters. Knowing a string's length can be very useful and is returned by the 'length' function.

```matlab
length(stringa)
```

You can extract a substring from a string by defining the index of the characters within curved brackets. Try the following commands.



```matlab
stringa(1)
stringa(1:5)
stringa(1:2:9)
```

Strings can also be converted to numbers - for example, each character has a numerical 'ascii' code



```matlab
asciistring = uint8(stringa)
```

They ascii codes can be converted back to text characters.



```matlab
char(asciistring)
```

Multi-line strings also are possible, however, each line must be the same length.



```matlab
stringb = ['a b c';'c d e']
```

We will learn more about string operations as we go along in later tutorials. For now, the most important thing is to understand the distinction between String and Number variables in MATLAB.
